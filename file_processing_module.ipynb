```python
# To read and write images, files has to be opened in binary mode
with open('harshil.jpg','rb') as ri:
    with open('harshil2.jpg','wb') as wi:
        for line in ri:
            wi.write(line)
```

```python
# about csv files
import csv
with open('Billdesk1.csv','r') as rd:   # default parameters expect comma as delimiter
    contents1=csv.reader(rd)
     next(contents1) # it will skip over first line
    for line in rd:
        print(line)
 
```

```python
with open('Billdesk1.csv','r') as r:
    f1=csv.reader(r)
    with open('Billdesk2.csv','w') as b:
        f2=csv.writer(b,delimiter='-')
        for line in f1:
            f2.writerow(line)
```
